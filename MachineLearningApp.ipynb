```python
for col in data.columns:
    data[col].fillna(data[col].value_counts().index[0])
```

```python
for c in data.columns:
    data[c]=data[c].fillna()
```

```python
import sklearn
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder=LabelEncoder()
```

```python
set(data['marital-status'])
```

```python
data['marital-status'].replace([' Divorced',' Married-AF-spouse',' Married-civ-spouse',' Married-spouse-absent',' Never-married',' Separated',' Widowed'],['Divorced','Married','Married','Married','not married','not married','not married'],inplace=True)
```

```python
set(data['marital-status'])
```

```python
data.head()
```

```python
categorical=['workclass','marital-status','occupation','relationship','race','sex','native-country']
```

```python
data['workclass']=labelencoder.fit_transform(data['workclass'].astype(str))
```

```python
for c in categorical:
    data[c]=labelencoder.fit_transform(data[c].astype(str))
```

```python
data.head()
```

```python
data.drop("fnlwgt",axis=1,inplace=True)
```

```python
data.head()
```

```python
df=pd.read_csv('adult.csv',names=columns,index_col=False)
```

```python
df.head()
```

```python
Y=df.iloc[:,-1]
```

```python
Y.dtype
```

```python
df['income']=labelencoder.fit_transform(df['income'])
```

```python
df['income']
```

```python
Y=df['income']
```

```python
X=data
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=100)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_clf_gini = sklearn.tree.DecisionTreeClassifier(criterion = "gini", random_state = 100,max_depth=5, min_samples_leaf=5)
dt_clf_gini.fit(x_train, y_train)
y_pred_gini = dt_clf_gini.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print ("Desicion Tree using Gini Index\nAccuracy is ", accuracy_score(y_test,y_pred_gini)*100 )
```

```python
import pickle
file=open("model.pkl",'wb')
pickle.dump(dt_clf_gini,file)
file.close()
```
